```python
first_name=input("Enter your first name in lowercase: ")
last_name=input("Enter your last name in lowercase: ")
print(f"{first_name.capitalize()} {last_name.capitalize()}")

poem_line=input("Type in the first line of the poem Raven by Edgar Allen Poe: ")
print(f" line length: {len(poem_line)}")
starting_index=int(input("Enter a starting number: "))
ending_index=int(input("Enter an ending number: "))
print(f" {poem_line[starting_index:ending_index]}")

temp="    Hello      "
print(temp)
print(f" without extra white space on left: {temp.lstrip()}")
print(f" without extra white space on right: {temp.rstrip()}")
print(f" without extra white space: {temp.strip()}")
print("We remove whitespaces because Python processes them as characters and cannot properly process the text if we are unaware of them.")
```
