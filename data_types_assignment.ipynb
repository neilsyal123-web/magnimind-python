```python
capital=1000
daily_growth=1.12
period=7
final_growth=daily_growth**period
result=capital*final_growth
print(result)


print('''When we buy bitcoin with {} USD at the beginning of the week,
we would earn {} USD at the end of the week, with an average gain of {}%.'''.format(capital,result,(daily_growth-1)*100))

fahrenheit=float(input("Enter the temperature in Fahrenheit: "))
celcius=(5/9)*(fahrenheit-32)
print(f"That temperature in Celcius: {celcius}")

number=input("Enter a three digit number: ")
dig1=int(number[0])
dig2=int(number[1])
dig3=int(number[2])
sum=dig1+dig2+dig3
print(f"The sum of digits in the number is {sum}")

first_side=float(input("Enter the first side length: "))
second_side=float(input("Enter the second side length: "))
hypotenuse=((first_side**2)+(second_side**2))**(0.5)
print(f"The length of the hypotenuse is {hypotenuse}")
```
